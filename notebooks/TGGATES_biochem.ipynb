```r
biochem.raw <- dat.merged[,names(dat.biochem)[10:30]]
```

```r
head(biochem.raw)
```

```r
biochem.imputed <- biochem.raw[,!apply(is.na(biochem.raw),2,any)]
dim(biochem.imputed)
```

```r
names(which(apply(is.na(biochem.imputed),2,any)))
```

```r
biochem.raw <- biochem.imputed
rownames(biochem.raw) <- dat.merged$CEL
```

```r
head(biochem.raw)
```

```r
write.csv(biochem.raw,file='data/TGGATES/cel/rat/biochem.csv')
```

```r
biochem.z <- t((t(biochem.raw)-apply(biochem.raw,2,mean))/apply(biochem.raw,2,sd))
```

```r
head(biochem.z)
```

```r
biochem.betas <- apply(biochem.z,2,function(x){
    mod <- lm(y~0+combn,data=data.frame(y=x,combn=meta$combn))
    mod$coefficients
})
```

```r
head(biochem.betas)
```

```r
write.csv(biochem.betas,file='data/TGGATES/cel/rat/biochem.scaled.csv')
```

```r
common_cels <- intersect(rownames(design.all),rownames(biochem.betas))
length(common_cels)
```

```r
design.all <- design.all[,rownames(biochem.betas)]
```

```r
dim(design.all)
```

```r
biochem.fitted <- do.call(rbind,lapply(colnames(design.all),function(i){
    biochem.z[design.all[,i]>0,] - biochem.betas[i,]
}))
head(biochem.fitted)
```

```r
write.csv(biochem.fitted,file='data/TGGATES/cel/rat/biochem.fitted.csv')
```
