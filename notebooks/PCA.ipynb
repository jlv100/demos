```r
set.seed(123)
x <- runif(100)-.5
y <- x+rnorm(100,0,.1)
```

```r
library(ggplot2)
```

```r
ggplot(data.frame(x=x,y=y)) + geom_point(aes(x,y)) + xlim(-1,1) + ylim(-1,1)
```

```r
pca <- prcomp(cbind(x,y))
```

```r
pca.dat <- as.data.frame(predict(pca,cbind(x,y)))
```

```r
ggplot(pca.dat) + geom_point(aes(PC1,PC2)) + xlim(-1,1) + ylim(-1,1)
```
